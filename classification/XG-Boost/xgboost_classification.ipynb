### Training and test dataset

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder()
```

```python
# categorical variables
x_cat = creditcard_df[['SEX', 'EDUCATION', 'MARRIAGE']]
x_cat = one_hot_encoder.fit_transform(x_cat).toarray()
x_cat
```

```python
x_cat.shape
```

```python
x_cat = pd.DataFrame(x_cat)
```

```python
creditcard_df.columns
```

```python
# numerical variables
x_numerical = creditcard_df.drop(['SEX', 'EDUCATION', 'AGE', 'MARRIAGE', 'default.payment.next.month'], axis=1, inplace=False)
x_numerical.head()
```

```python
x = pd.concat([x_cat, x_numerical], axis=1)
```

```python
y = creditcard_df['default.payment.next.month']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### Model

```python
import xgboost as xgb
model = xgb.XGBClassifier(learning_rate=0.1, 
                          max_depth=20,
                          n_estimators=500,
                          use_label_encoder=False)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_predict = model.predict(x_test)
```

```python
print('Accuracy:', 100 * accuracy_score(y_predict, y_test))
```

```python
cm = confusion_matrix(y_predict, y_test)
cm
```

```python
sns.heatmap(cm, annot=True, fmt='d')
```

```python
print(classification_report(y_test, y_predict))
```
